```python
tcga_full = tcga_full.dropna()
tcga_full
```

```python
# filter genes
partner_genes_to_filter = pd.read_csv('/Users/andrei/Documents/GitHub/cancer-low-data/data/imp_genes_list.csv',index_col=0)
set_partner_genes_to_filter = set(partner_genes_to_filter.index)
set_partner_genes_to_filter
```

```python
tumor_grade = tcga_full['tumor_grade']
tumor_grade
```

```python
two_year_survival = tcga_full['X2yr.RF.Surv.']
two_year_survival
```

```python
# Clean the set of columns to match the Partner Naming
column_names_clean = []
for column_with_number in tcga_full.columns:
	column_name = column_with_number.split('|')[0]
	column_names_clean.append(column_name)

column_names_clean
```

```python
# keep only the Partner set of genes
tcga_full_columns_changed = tcga_full.copy()
tcga_full_columns_changed.columns = column_names_clean
```

```python
# print the number of shared genes between the TCGA dataset and Partner
genes_intersection = list(set(column_names_clean).intersection(set_partner_genes_to_filter))
len(genes_intersection)
```

```python
tcga_only_intersection_genes = tcga_full_columns_changed[genes_intersection]
tcga_only_intersection_genes.shape
```

# Make tumor grade splits

```python
tumor_grade.value_counts()
```

```python
# Decide to keep only G1, G2, G3 classes
tcga_genes_and_tumor_grade = tcga_only_intersection_genes.copy()
tcga_genes_and_tumor_grade = tcga_genes_and_tumor_grade.merge(tumor_grade, left_index=True, right_index=True, validate='one_to_one')
tcga_genes_and_tumor_grade = tcga_genes_and_tumor_grade.loc[tcga_genes_and_tumor_grade['tumor_grade'].isin(['G1', 'G2', 'G3'])]
tcga_genes_and_tumor_grade.shape
```

```python
tcga_genes_and_tumor_grade['tumor_grade'] = tcga_genes_and_tumor_grade['tumor_grade'].map({'G1': int(0), 'G2': int(1), 'G3': int(2)})
tcga_genes_and_tumor_grade['tumor_grade'].value_counts()
```

```python
# create test set
data_tumor_grade_train_val, data_tumor_grade_test = train_test_split(tcga_genes_and_tumor_grade, test_size=100, random_state=42, shuffle=True, stratify=tcga_genes_and_tumor_grade['tumor_grade'])
data_tumor_grade_test.to_csv(f"/Users/andrei/Documents/GitHub/cancer-low-data/data/TCGA_samples/tcga_tumor_grade_test_100.csv")
```

```python
# Create validation set
data_tumor_grade_train, data_tumor_grade_val = train_test_split(data_tumor_grade_train_val, test_size=50, random_state=42, shuffle=True, stratify=data_tumor_grade_train_val['tumor_grade'])
data_tumor_grade_val.to_csv(f"/Users/andrei/Documents/GitHub/cancer-low-data/data/TCGA_samples/tcga_tumor_grade_val_50.csv")
```

```python
# Make train sets splits
for train_size in [100, 200, 330]:
	tumor_grade_cohort, _ = train_test_split(data_tumor_grade_train, train_size=train_size, random_state=42, shuffle=True, stratify=data_tumor_grade_train['tumor_grade'])
	tumor_grade_cohort.to_csv(f"/Users/andrei/Documents/GitHub/cancer-low-data/data/TCGA_samples/tcga_tumor_grade_train_{train_size}.csv")

	print(f"Cohort size {len(tumor_grade_cohort)}")
	print(tumor_grade_cohort['tumor_grade'].value_counts())
```

# Make 2 year RF

```python
two_year_survival.value_counts()
```

```python
tcga_genes_and_2ysurvival = tcga_only_intersection_genes.copy()
tcga_genes_and_2ysurvival = tcga_genes_and_2ysurvival.merge(two_year_survival, left_index=True, right_index=True, validate='one_to_one')
tcga_genes_and_2ysurvival.shape
```

```python
# Create test set
data_2ysurvival_train_val, data_2ysurvival_test = train_test_split(tcga_genes_and_2ysurvival, test_size=100, random_state=42, shuffle=True, stratify=tcga_genes_and_2ysurvival['X2yr.RF.Surv.'])
data_2ysurvival_test.to_csv(f"/Users/andrei/Documents/GitHub/cancer-low-data/data/TCGA_samples/tcga_2ysurvival_test_100.csv")
```

```python
# Create validation set
data_2ysurvival_train, data_2ysurvival_val = train_test_split(data_2ysurvival_train_val, test_size=50, random_state=42, shuffle=True, stratify=data_2ysurvival_train_val['X2yr.RF.Surv.'])
data_2ysurvival_val.to_csv(f"/Users/andrei/Documents/GitHub/cancer-low-data/data/TCGA_samples/tcga_2ysurvival_val_50.csv")
```

```python
# Make splits
for train_size in [100, 200, 330]:
	two_year_survival_cohort, _ = train_test_split(data_2ysurvival_train, train_size=train_size, random_state=42, shuffle=True, stratify=data_2ysurvival_train['X2yr.RF.Surv.'])
	two_year_survival_cohort.to_csv(f"/Users/andrei/Documents/GitHub/cancer-low-data/data/TCGA_samples/tcga_2ysurvival_train_{train_size}.csv")

	print(f"Cohort size {len(two_year_survival_cohort)}")
	print(two_year_survival_cohort['X2yr.RF.Surv.'].value_counts())
```
